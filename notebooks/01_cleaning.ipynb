```python
df['top_challenge'] = df['top_challenge'].fillna('Unknown')
```

```python
print('=== NULL VALUES ===')
print(df.isnull().sum())
```

# Remove extra spaces and lowercase all string columns


```python
cat_cols = ['gender', 'region', 'occupation', 'education_level',
'preferred_product_category', 'purchase_channel', 'top_challenge']
```

```python
for col in cat_cols:
 df[col] = df[col].str.strip().str.lower()
```

# Check unique values

```python
for col in cat_cols:
  print(f'{col}: {df[col].unique()}', "\n")
```

```python
# Monthly spend = visits x avg spend per visit
df['monthly_spend_birr'] = df['visits_per_month'] * df['avg_spend_per_visit_birr']
# Customer value score: normalized spend x satisfaction
df['spend_norm'] = (df['monthly_spend_birr'] - df['monthly_spend_birr'].min()) / \
(df['monthly_spend_birr'].max() - df['monthly_spend_birr'].min())
df['sat_norm'] = df['satisfaction_score'] / 5
df['customer_value_score'] = (df['spend_norm'] * 0.6) + (df['sat_norm'] * 0.4)
print(df[['monthly_spend_birr','customer_value_score']].describe())
```

```python
df.to_csv('../data/processed/survey_cleaned.csv', index=False)
print(f'Saved: {df.shape[0]} rows, {df.shape[1]} columns')
print(df.columns.tolist())
```
